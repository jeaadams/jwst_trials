```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

### This notebook is meant to help analyse how changing algorithm aggressiveness by using different KLIP parameters affects our contrast curves

```python
df = pd.read_csv('contrast_meas.csv')
```

```python
ax = sns.relplot(x="separation", y="med_contrast", palette = 'Pastel1', height = 3.5, aspect = 1.5, col_wrap = 2, col = 'minrot', hue="kl", lw=1, data=df, kind = 'line')
plt.yscale('log')
ax.set_xlabels('Separation')
ax.set_ylabels('Median Contrast')
ax.savefig('kl_med_contrast.png', dpic = 500)
```

```python
ax = sns.relplot(x="separation", y="smooth_contrast", palette = 'Pastel1', height = 3.5, aspect = 1.5, col_wrap = 2, col = 'minrot', hue="kl", lw=1, data=df, kind = 'line')
plt.yscale('log')
ax.set_xlabels('Separation')
ax.set_ylabels('Best Fit Contrast')
ax.savefig('kl_best_contrast.png', dpic = 500)
```

```python
ax = sns.relplot(x="separation", y="smooth_contrast", palette= 'colorblind', height = 3.5, aspect = 1.5, col_wrap = 2, col = 'kl', hue="minrot", lw=1, data=df, kind = 'line')
plt.yscale('log')
ax.set_xlabels('Separation')
ax.set_ylabels('Best Fit Contrast')
ax.savefig('minrot_best_contrast.png', dpic = 500)
```

```python
ax = sns.relplot(x="separation", y="med_contrast", palette= 'colorblind', height = 3.5, aspect = 1.5, col_wrap = 2, col = 'kl', hue="minrot", lw=1, data=df, kind = 'line')
ax.set_xlabels('Separation')
ax.set_ylabels('Median Contrast')
plt.yscale('log')
ax.savefig('minrot_med_contrast.png', dpic = 500)
```

```python
ax = sns.relplot(x="separation", y="med_contrast",  palette = 'Dark2', hue="kl", height = 5, aspect = 2, style = 'minrot', lw=1, data=df, kind = 'line')
ax.set_xlabels('Separation')
ax.set_ylabels('Median Contrast')
plt.yscale('log')
ax.savefig('all_params.png', dpic = 500)
```
